```python
# Define the relu function
def relu(x):
    if x > 0:
        return x
    else:
        return 0

numbers = [1,-2,3,-5]
result = [relu(num) for num in numbers]
print("Original Numbers:", numbers)
print("Numbers after applying ReLU:", result)
```

```python
import random

# Define the relu function
def relu(x):
    if x > 0:
        return x
    else:
        return 0

# Generate 10 random integers including positive and negative values
numbers = [random.randint(-10, 10) for _ in range(10)]

# Apply relu function to each number
result = [relu(num) for num in numbers]

print("Original Numbers:", numbers)
print("Numbers after applying ReLU:", result)
```

```python
import matplotlib.pyplot as plt


# Plotting
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.bar(range(len(numbers)), numbers, color='b')
plt.title('Original Numbers')
plt.xlabel('Index')
plt.ylabel('Value')

plt.subplot(2, 1, 2)
plt.bar(range(len(result)), result, color='r')
plt.title('Numbers after applying ReLU')
plt.xlabel('Index')
plt.ylabel('Value')

plt.tight_layout()
plt.show()
```
